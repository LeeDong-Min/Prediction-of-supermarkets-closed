```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,GridSearchCV
```

```python
emart = pd.read_excel("./prototype_model3.xlsx",usecols="A,I,J")
```

```python
emart
```

```python
x = emart.iloc[:,-1]
y = emart.iloc[:,1]
```

```python
x.shape
```

```python
y.shape
```

```python
X = np.array(x).reshape(-1,1)
```

```python
X.shape
```

```python
LR = LogisticRegression()
```

```python
LR.fit(X,y)
```

```python
LR.score(X,y)
```

```python
LR.predict(X)
```

```python
pred=LR.predict_proba(X)
```

```python
for i in range(len(emart)):
    if pred[i][1] > 0.2:
        print(emart["지점명"][i],"----> 생존확률:{} / 폐점확률:{} ---->{}".format(pred[i][0],pred[i][1],"폐점가능성"))
```

```python
grid={"C" : np.logspace(0, 4, 10), "penalty":["l1","l2"],"solver":["newton-cg", "lbfgs", "liblinear", "sag", "saga"]}
```

```python
logreg_cv=GridSearchCV(LR,grid,cv=10)
```

```python
logreg_cv.get_params
```

```python
logreg_cv.fit(X,y)
```

```python
logreg_cv.predict(X)
```

```python
pred=logreg_cv.predict_proba(X)
```

```python
voting = pd.DataFrame()
```

```python
voting["지점명"] = emart["지점명"]
voting["3차폐점확률"] = 0.
for i in range(len(emart)):
    if pred[i][1] > 0.2:
        print(emart["지점명"][i],"----> 생존확률:{} / 폐점확률:{} ---->{}".format(pred[i][0],pred[i][1],"폐점가능성"))
        voting.loc[i,"3차폐점확률"] = pred[i][1]
```

```python
voting.to_excel("make_voting.xlsx",index=False)
```
