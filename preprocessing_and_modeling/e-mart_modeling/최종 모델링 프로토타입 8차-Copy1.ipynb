```python
emart = pd.read_excel("../../데이터/emart(최종.ver1).xlsx",usecols="A,C,F,T,V:W,Z:AB")
emart
```

```python
emart = emart[emart.점포형태 != 1]
emart.reset_index(drop=True, inplace = True)
```

```python
emart.corr()
```

```python
df_close = emart[emart.점포형태 == 2]
```

```python
columns_li=['1인당지방세부담액(천원)', '다인가구구성원수', '카페(1년)', '등록자동차대수']

for col in columns_li:
    _min=round(df_close[col].mean()*0.8,3)
    _max=round(df_close[col].mean()*1.2,3)
    print(col)
    print('평균값:',df_close[col].mean())
    print('최고값:',_max, '최저값:',_min)
    def df_value(row):
        if row<=_min:
            return 3
        elif (row>_min and row<=_max):
            return 2
        elif row>_max:
            return 1
    emart[col]=emart[col].apply(df_value)
```

```python
#하 중 상
columns_li=['3km이마트','3km경쟁마트', '1인가구구성원수']
for col in columns_li:
    _min=round(df_close[col].mean()*0.8,3)
    _max=round(df_close[col].mean()*1.2,3)
    print(col)
    print('평균값:',df_close[col].mean())
    print('최고값:',_max, '최저값:',_min)
    def df_value2(row):
        if row<=_min:
            return 1
        elif (row>_min and row<=_max):
            return 2
        elif row>_max:
            return 3
    emart[col]=emart[col].apply(df_value2)
```
