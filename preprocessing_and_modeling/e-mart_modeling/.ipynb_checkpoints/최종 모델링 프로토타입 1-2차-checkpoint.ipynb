```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,GridSearchCV
```

```python
emart = pd.read_excel("./prototype_model.xlsx",usecols="A,K,L")
```

```python
emart
```

```python
x = emart.iloc[:,1]
y = emart.iloc[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
X = np.array(x).reshape(-1,1)
```

```python
X.shape
```

```python
LR = LogisticRegression()
```

```python
LR.fit(X,y)
```

```python
LR.score(X,y)
```

```python
LR.predict(X)
```

```python
pred=LR.predict_proba(X)
```

```python
for i in range(len(emart)):
    if pred[i][1] > pred[i][0]:
        print(emart["지점명"][i],"----> 생존확률:{} / 폐점확률:{} ---->{}".format(pred[i][0],pred[i][1],"폐점확정"))
    else:
        print(emart["지점명"][i],"----> 생존확률:{} / 폐점확률:{} ---->{}".format(pred[i][0],pred[i][1],"생존확정"))
```
