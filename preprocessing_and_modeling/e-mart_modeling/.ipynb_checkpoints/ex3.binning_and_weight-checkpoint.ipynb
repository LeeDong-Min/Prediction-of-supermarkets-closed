```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_excel("./prototype_model_3mart.xlsx")
test = pd.read_excel("./emart_binning_model.xlsx")
```

```python
train
```

```python
test
```

```python
train.corr()
```

```python
for i in range(len(train)):
    train.iloc[i,2] = train.iloc[i,2]*3.21
    train.iloc[i,3] = train.iloc[i,3]*2.29
    train.iloc[i,4] = train.iloc[i,4]*1.12
    train.iloc[i,5] = train.iloc[i,5]*0.8
    train.iloc[i,6] = train.iloc[i,6]*0.8
    train.iloc[i,7] = train.iloc[i,7]*3.35
    train.iloc[i,8] = train.iloc[i,8]*1.3
    
for i in range(len(test)):
    test.iloc[i,1] = test.iloc[i,1]*3.21
    test.iloc[i,2] = test.iloc[i,2]*2.29
    test.iloc[i,3] = test.iloc[i,3]*1.12
    test.iloc[i,4] = test.iloc[i,4]*0.8
    test.iloc[i,5] = test.iloc[i,5]*0.8
    test.iloc[i,6] = test.iloc[i,6]*3.35
    test.iloc[i,7] = test.iloc[i,7]*1.3
```

```python
train
```

```python
train_X = train.iloc[:,2:9]
train_y = train.iloc[:,-1]
```

```python
test_X = test.iloc[:,1:8]
test_y = test.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(train_X,train_y)
```

```python
LR.score(test_X,test_y)
```

```python
LR.predict(test_X)
```

```python
voting = pd.read_excel("./voting2.xlsx")
voting
```

```python
pred = LR.predict_proba(test_X)
```

```python
voting["binningAndWeight"] = 0.
for i in range(len(test)):
    if pred[i][0] >= 0.3:
        print(test["지점명"][i],"----> 생존확률:{} / 폐점확률:{} ---->{}".format(pred[i][1],pred[i][0],"폐점가능성"))
        voting.iloc[voting[voting.지점명 == test["지점명"][i]].index[0],-1] = pred[i][0]
```

```python
voting
```

```python
voting.to_excel("voting2.xlsx",index=False)
```
