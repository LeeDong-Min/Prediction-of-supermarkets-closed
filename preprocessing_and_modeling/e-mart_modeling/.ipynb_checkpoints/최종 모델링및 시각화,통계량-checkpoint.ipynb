```python
import numpy as np
import pandas as pd
import seaborn as sns
import pandas_profiling
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
sns.set(style='darkgrid')


rc('font', family='AppleGothic')
matplotlib.rcParams['axes.unicode_minus'] = False
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
emart = pd.read_excel("./final_model.xlsx")
emart
```

```python
for i in range(len(emart)):
    if emart.loc[i,"3km내 자사"] == 0:
        emart.loc[i,"3km내 자사"] = 1
    elif emart.loc[i,"3km내 자사"] == 1:
        emart.loc[i,"3km내 자사"] = 2
    else:
        emart.loc[i,"3km내 자사"] = 3
```

```python
for i in range(len(emart)):
    if emart.loc[i,"3km내 경쟁"] == 0:
        emart.loc[i,"3km내 경쟁"] = 1
    elif emart.loc[i,"3km내 경쟁"] == 1:
        emart.loc[i,"3km내 경쟁"] = 2
    else:
        emart.loc[i,"3km내 경쟁"] = 3
```

```python
emart
```

```python
emart_a = emart[emart["점포형태"]==1].sample(41,random_state = 77)
```

```python
emart_a.reset_index(drop=True,inplace=True)
```

```python
emart_b =emart[emart["점포형태"]==0]
```

```python
emart_c=pd.concat([emart_a,emart_b])
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.preprocessing import robust_scale

LR = LogisticRegression()

X = emart_c.iloc[:,1:-1]
y = emart_c.iloc[:,-1]

X = robust_scale(X)
```

```python
LR.fit(X,y)
```

```python
LR.predict(robust_scale(emart.iloc[:,1:-1]))
```

```python
for i,v in enumerate(np.round(LR.predict_proba(robust_scale(emart.iloc[:,1:-1])),3)):
    print(emart.iloc[i,0],v)
```

```python
for i,v in enumerate(np.round(LR.predict_proba(robust_scale(emart.iloc[:,1:-1])),3)):
    if v[0] > 0.5:
        print(emart.iloc[i,0],v)
```

# 통계량

```python
import statistics
from scipy import stats

def get_pvalue(pop, sam, col):
    x_bar = sam[col].mean()
    mu = pop[col].mean()
    sigma = statistics.stdev(pop[col])
    n = 50
    
    print("'"+col+"'의 통계")

    t_sample = (x_bar - mu) /(sigma / np.sqrt(float(n)))
    print("검정 통계량: ",np.round(t_sample, 2))

    p_val = stats.t.sf(np.abs(t_sample), n-1)
    print("t-table의 아래폭 꼬리 p 값: ",np.round(p_val, 4))
```

```python
for i in emart.keys()[1:-1]:
    get_pvalue(emart.iloc[:,1:], emart_c.iloc[:,1:], i)
    print("-"*100)
```
