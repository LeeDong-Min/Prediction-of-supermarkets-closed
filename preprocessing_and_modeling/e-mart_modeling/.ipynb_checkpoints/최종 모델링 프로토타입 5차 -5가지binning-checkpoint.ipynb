```python
for i in range(len(emart)):
    if emart.loc[i,"1인당지방세부담액(천원)"]//100 < 9:
        emart.loc[i,"1인당지방세부담액(천원)"] = 5
    elif emart.loc[i,"1인당지방세부담액(천원)"]//100 < 12:
        emart.loc[i,"1인당지방세부담액(천원)"] = 4
    elif emart.loc[i,"1인당지방세부담액(천원)"]//100 < 18:
        emart.loc[i,"1인당지방세부담액(천원)"] = 3
    elif emart.loc[i,"1인당지방세부담액(천원)"]//100 < 30:
        emart.loc[i,"1인당지방세부담액(천원)"] = 2
    else:
        emart.loc[i,"1인당지방세부담액(천원)"] = 1
```
