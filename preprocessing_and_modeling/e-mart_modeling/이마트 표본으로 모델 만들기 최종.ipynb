```python
emart_a.reset_index(drop=True,inplace=True)
```

```python
emart_c=pd.concat([emart_a,emart_b])
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.preprocessing import scale, robust_scale

LR = LogisticRegression()

X = emart_c.iloc[:,1:-1]
y = emart_c.iloc[:,-1]

X.info()

X1 = scale(X)
X2 = robust_scale(X)
```

```python
LR.fit(X1,y)

a = np.round(LR.predict_proba(X1),2)

print(LR.predict(scale(emart.iloc[:,1:-1])))

LR.fit(X2,y)

b =np.round(LR.predict_proba(X2))

print(LR.predict(robust_scale(emart.iloc[:,1:-1])))
```
