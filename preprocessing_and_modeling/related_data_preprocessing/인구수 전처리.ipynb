```python
import numpy as np
import pandas as pd
```

```python
population = pd.read_excel("../데이터/인구수/구별 인구수.xlsx",header=1)
```

```python
population.fillna(method="pad", inplace = True)
```

```python
population.rename(columns= {"행정구역(동읍면)별":"광역시도",
                           "항목":'시자치구',
                           '계':'인구수'},inplace=True)

# population = population[(population['시자치구'] != '소계')]
```

```python
population.is_copy = False

population.rename(columns= {"항목.1":"구분"},inplace=True)


population.loc[population["구분"]=="총인구수 (명)","구분"] = "합계"
population.loc[population["구분"]=="남자인구수 (명)","구분"] = "남자"
population.loc[population["구분"]=="여자인구수 (명)","구분"] = "여자"
```
