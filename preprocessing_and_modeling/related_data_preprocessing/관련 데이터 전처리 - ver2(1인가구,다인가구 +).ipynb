```python
one_fam = pd.read_excel("../데이터/관련자료/1인가구비율_시도_시_군_구__20200628203653.xlsx",header=1,usecols="A:B,F:H")
```

```python
one_fam
```

```python
# 소계 다날리고 셰종은 안합치고
one_fam = one_fam[one_fam["행정구역별.1"] != "소계"]
```

```python
one_fam.reset_index(drop=True, inplace=True)
```

```python
one_fam["항목"] =""
```

```python
for i in range(len(one_fam["행정구역별"])):
    if one_fam.loc[i,"행정구역별.1"] != one_fam.loc[i,"행정구역별"]:
        if "특별시" in one_fam.loc[i,"행정구역별"] or "광역시" in one_fam.loc[i,"행정구역별"]:
            one_fam.loc[i,"항목"] = one_fam.loc[i,"행정구역별"] + " " + one_fam.loc[i,"행정구역별.1"].strip()
        else:
            one_fam.loc[i,"항목"] = one_fam.loc[i,"행정구역별.1"].strip()
    else:
        one_fam.loc[i,"항목"] = one_fam.loc[i,"행정구역별.1"].strip()
```

```python
one_fam
```

```python
pop["1인가구구성원수"] = 0
pop["다인가구구성원수"] = 0
```

```python
pop
```

```python
for i in range(len(one_fam["항목"])):
    for j in range(len(pop["세부사항"])):
        if one_fam.loc[i,"항목"] in pop.loc[j,"세부사항"]:
            pop.loc[j,"1인가구구성원수"] = one_fam.loc[i,"1인가구구성원수"]
            pop.loc[j,"다인가구구성원수"] = one_fam.loc[i,"다인가구구성원수"]
```
