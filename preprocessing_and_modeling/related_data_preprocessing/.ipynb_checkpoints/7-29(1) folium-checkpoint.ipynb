```python
emart=pd.read_excel('../../데이터/이마트기본데이터/이마트 데이터.xlsx',usecols="B,D,E")
```

```python
import os
from folium import MarkerCluster
```

```python
m=folium.Map(location=[36.433271, 128.058813],zoom_start=6, tiles='OpenStreetMap')
mc = plugins.MarkerCluster() 

for i in emart.index:
    mc.add_child(folium.Marker(location=[emart.loc[i,'위도'],emart.loc[i,"경도"]],icon=folium.Icon(color='red',icon='ok')))

#     folium.CircleMarker(location=[emart.loc[i,'위도'],emart.loc[i,"경도"]],radius=5,color='#ffa500',fill=True).add_to(m)
    

m.add_child(mc)
m.save('emart_folium.html')
m
```
