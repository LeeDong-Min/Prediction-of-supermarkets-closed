```python
import numpy as np
import pandas as pd
```

```python
population = pd.read_excel("../데이터/인구수/구별 인구수.xlsx",header=1)
```

```python
population
```

```python
population.fillna(method="pad", inplace = True)
```

```python
population
```

```python
population.rename(columns= {"행정구역(동읍면)별":"광역시도",
                           "항목":'시자치구',
                           '계':'인구수'},inplace=True)

population = population[(population['시자치구'] != '소계')]
```

```python
population
```

```python
population.is_copy = False

population.rename(columns= {"항목.1":"구분"},inplace=True)


population.loc[population["구분"]=="총인구수 (명)","구분"] = "합계"
population.loc[population["구분"]=="남자인구수 (명)","구분"] = "남자"
population.loc[population["구분"]=="여자인구수 (명)","구분"] = "여자"
```

```python
population
```

```python
population["20 - 39세"] = population.loc[:,"20 - 24세":"35 - 39세"].sum(1)
population["65세이상"] = population.loc[:,"65 - 69세":"100+"].sum(1)
```

```python
pop = pd.pivot_table(population,
                    index=["광역시도","시자치구"],
                    columns =["구분"],
                    values = ["20 - 39세","65세이상","인구수"])
pop
```

```python
pop['소멸비율'] = pop['20 - 39세','여자'] / (pop['65세이상','합계'] / 2)
pop.head(10)
```

```python
pop['소멸위기지역'] = pop['소멸비율'] < 0.5
pop.head(10)
```

```python
pop[pop["소멸위기지역"]==True].index.get_level_values(1)
```

```python
pop.reset_index(inplace=True) #인덱스가 다시 숫자로
pop.head()
```

```python
tmp_coloums = [pop.columns.get_level_values(0)[n] + pop.columns.get_level_values(1)[n]
              for n in range(0,len(pop.columns.get_level_values(0)))]

pop.columns = tmp_coloums

pop.head()
```

```python
pop.info()
```

```python
new_pop = pop.iloc[:,[1,-2,-1]]
```

```python
new_pop.to_excel("인구소멸 지역.xlsx",index=False)
```
