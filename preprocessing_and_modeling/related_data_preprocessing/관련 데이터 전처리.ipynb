```python
import numpy as np
import pandas as pd
```

```python
pop = pd.read_excel("../데이터/관련자료/구별 인구수.xlsx",header=1,usecols="A:D,R:Y")
pop["65세이상"] = pop.loc[:,"65 - 69세":"100+"].sum(1)
```

```python
pop.fillna(method="pad",inplace=True)
```

```python
pop
```

```python
pop = pop[["행정구역(동읍면)별","항목","항목.1","계","65세이상"]]
```

```python
pop
```

```python
pop = pop[pop["항목.1"]=="총인구수 (명)"]
```

```python
pop
```

```python
pop = pop[pop["항목"] != "소계"][["행정구역(동읍면)별","항목","65세이상","계"]]
```

```python
pop.reset_index(drop=True,inplace=True)
```

```python
pop["항목(2)"] = ""
```

```python
pop
```

```python
for i in range(len(pop["행정구역(동읍면)별"])):
    if len(pop.loc[i,"항목"]) < 8:
        pop.loc[i,"항목(2)"] = pop.loc[i,"행정구역(동읍면)별"] + " " + pop.loc[i,"항목"]
    else:
        pop.loc[i,"항목(2)"] = pop.loc[i,"항목"]
```

```python
pop
```

```python
tax = pd.read_excel("../데이터/관련자료/1인당부담액.xlsx",usecols="B,C,D,E")
tax
```

```python
tax = tax[tax["행정구역별(2)"] != "소계"]
```

```python
tax = tax[tax["행정구역별(2)"] != "총계"]
```

```python
tax.reset_index(drop=True,inplace=True)
tax
```

```python
pop["1인당부담액 (천원)"] = 0
```

```python
pop
```

```python
for i in range(len(tax["행정구역별"])):
    for j in range(len(pop["항목(2)"])):
        if tax.loc[i,"행정구역별"] in pop.loc[j,"항목(2)"]:
            pop.loc[j,"1인당부담액 (천원)"] = tax.loc[i,"1인당부담액 (천원)"]
```

```python
exit = pd.read_excel("../데이터/관련자료/인구소멸 지역.xlsx")
exit
```

```python
pop["소멸위기지역"] = 0
```

```python
pop
```

```python
for i in range(len(exit["시자치구"])):
    for j in range(len(pop["항목"])):
        if exit.loc[i,"시자치구"] == pop.loc[j,"항목"]:
            if exit.loc[i,"소멸위기지역"]:
                pop.loc[j,"소멸위기지역"] = 1
            else:
                pop.loc[j,"소멸위기지역"] = 0
            
```

```python
pop
```

```python
pop["65세이상경제활동인구"] = 0
```

```python
pop
```

```python
work65 = pd.read_excel("../데이터/관련자료/65세이상_경제활동인구.xlsx")
work65
```

```python
for i in range(len(work65.시도별)):
    for j in range(len(pop["행정구역(동읍면)별"])):
        if work65.loc[i,"시도별"] == pop.loc[j,"행정구역(동읍면)별"]:
            pop.loc[j,"65세이상경제활동인구"] = work65.loc[i,"경제활동참가율(%)"] * pop.loc[j,"65세이상"] * 0.01
```

```python
pop
```

```python
work = pd.read_excel("../데이터/관련자료/시도별_경제활동인구비율.xlsx",header=1)
work
```

```python
pop["경제활동인구"] = 0
```

```python
pop
```

```python
for i in range(len(work.시도별)):
    for j in range(len(pop["행정구역(동읍면)별"])):
        if work.loc[i,"시도별"] == pop.loc[j,"행정구역(동읍면)별"]:
            pop.loc[j,"경제활동인구"] = work.loc[i,"경제활동참가율 (%)"] * pop.loc[j,"계"] * 0.01
```
