```python
import pandas as pd
```

```python
emart = pd.read_excel("../TeamProjectData/이마트 데이터.xlsx")
emart.지역.fillna(method='pad',inplace=True)
```

```python
emart_branch = emart.지점명
```

```python
emart_branch
```

```python
gmaps.geocode("이마트 청계천점",language="ko")
```

```python
emart_branch_name = ["이마트 "+ name  for name in emart_branch]
emart_branch_name
```

```python
emart_address = []
emart_lat = []
emart_lng = []

for name in emart_branch_name:
    tmp = gmaps.geocode(name, language = "ko")
    emart_address.append(tmp[0].get("formatted_address"))
    
    tmp_loc = tmp[0].get("geometry")
    
    emart_lat.append(tmp_loc["location"]["lat"])
    
    emart_lng.append(tmp_loc["location"]["lng"])
    
    print(name + "--->" + tmp[0].get("formatted_address"))
```

```python
emart.위도 = emart_lat
emart.경도 = emart_lng
```

```python
ts_strftime = [emart["개점 연도"][i].strftime('%Y-%m-%d') for i in range(len(emart["개점 연도"]))]
```

```python
ts_strftime
```

```python
emart["개점 연도"] = ts_strftime
```
