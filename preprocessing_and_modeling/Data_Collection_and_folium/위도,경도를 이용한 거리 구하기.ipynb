```python
import numpy as np
import pandas as pd
import math
```

```python
emart = pd.read_excel("../../데이터/이마트기본데이터/이마트 데이터.xlsx",usecols="B,D,E")
emart
```

```python
from haversine import haversine
```

```python
count_3, count_5, count_9 = [], [], []
for i in range(len(emart)):
    count3 = 0
    count5 = 0
    count9 = 0
    lat1 = emart.loc[i,"위도"]
    lon1 = emart.loc[i,"경도"]
    for j in range(len(emart)):
        if emart.loc[j, "지점명"] == emart.loc[i, "지점명"]:
            continue
        lat2 = emart.loc[j,"위도"]
        lon2 = emart.loc[j,"경도"]
        mart1 = (lat1, lon1) # (lat, lon)
        mart2 = (lat2, lon2)

        distance = haversine(mart1, mart2)
        if distance <= 3:
            count3 += 1
        if distance <= 5:
            count5 += 1
        if distance <= 9:
            count9 += 1
    
    count_3.append(count3)
    count_5.append(count5)
    count_9.append(count9)
```

```python
emart["3km이마트"] = count_3
emart["5km이마트"] = count_5
emart["9km이마트"] = count_9
```
